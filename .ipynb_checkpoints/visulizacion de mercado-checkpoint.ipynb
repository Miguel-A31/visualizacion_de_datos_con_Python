```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_supermarket_sales = pd.read_csv('supermarket_sales.csv')
df_supermarket_sales.head()
```

```python
df_supermarket_sales.info()
```

```python
df_supermarket_sales.isnull().sum()
```

```python
# el valor de gm_pct es el poncentaje de ganacia que se repite y no varia. No proporciona informacion relevante por ello se elimina
df_supermarket_sales['gm_pct'].value_counts()
```

```python
df_supermarket_sales = df_supermarket_sales.drop(columns=['gm_pct'])
```

```python
df_supermarket_sales.corr()
```

vemos que hay buena correlaci√≥n entre los datos

```python
table = pd.crosstab(df_supermarket_sales['gender_customer'], df_supermarket_sales['product_line'], normalize='index') * 100
table = table.round(2).astype(str) + '%'
table
```

```python
# Crear un gr√°fico de barras
sns.countplot(x='gender_customer', hue='product_line', data=df_supermarket_sales)

# A√±adir etiquetas y t√≠tulo
plt.xlabel('G√©nero del Cliente')
plt.ylabel('N√∫mero de Ventas')
plt.title('Distribuci√≥n de Ventas por G√©nero del Cliente y L√≠nea de Producto')

# Mover la leyenda fuera del gr√°fico
plt.legend(title='L√≠nea de Producto', bbox_to_anchor=(1, 1), loc='upper left')

# Mostrar el gr√°fico
plt.show()
```

```python
# Crear un gr√°fico de barras apiladas
sns.countplot(x='city', hue='product_line', data=df_supermarket_sales)

# A√±adir etiquetas y t√≠tulo
plt.xlabel('Ciudad')
plt.ylabel('N√∫mero de Ventas')
plt.title('Distribuci√≥n de Ventas por Ciudad y L√≠nea de Producto')

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45, ha='right')

# Ajustar la leyenda fuera del gr√°fico
plt.legend(title='L√≠nea de Producto', bbox_to_anchor=(1, 1), loc='upper left')

# Mostrar el gr√°fico
plt.show()
```

```python
# Crear un gr√°fico de barras
sns.set(style="whitegrid")
ax = sns.countplot(x='city', hue='gender_customer', data=df_supermarket_sales)

# A√±adir etiquetas y t√≠tulo
plt.xlabel('Ciudad')
plt.ylabel('N√∫mero de Ventas')
plt.title('Distribuci√≥n de G√©nero de Clientes por Ciudad')

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45, ha='right')

# Ajustar la leyenda fuera del gr√°fico
plt.legend(title='G√©nero', bbox_to_anchor=(1, 1), loc='upper left')

# Mostrar la tabla en porcentaje
table = pd.crosstab(df_supermarket_sales['city'], df_supermarket_sales['gender_customer'], normalize='index') * 100
table = table.round(2).astype(str) + '%'

# Imprimir la tabla
print(table)
plt.show()
```

```python
# Crear una tabla de frecuencia cruzada
cross_tab = pd.crosstab(index=[df_supermarket_sales['city'], df_supermarket_sales['gender_customer']],
                        columns=df_supermarket_sales['product_line'],
                        normalize='index') * 100

# Redondear los valores y convertir a porcentaje
cross_tab = cross_tab.round(2).astype(str) + '%'

# Imprimir la tabla
cross_tab
```

```python
# Crear un gr√°fico de barras agrupadas
sns.catplot(x='city', hue='gender_customer', col='product_line', kind='count', data=df_supermarket_sales, height=4, aspect=0.8)

# A√±adir etiquetas y t√≠tulo
plt.subplots_adjust(top=0.85)
plt.suptitle('Distribuci√≥n de G√©nero de Clientes por Ciudad y L√≠nea de Producto', fontsize=20)
plt.xlabel('Ciudad')
plt.ylabel('N√∫mero de Ventas')

# Mostrar el gr√°fico
plt.show()
```

```python
# Crear un gr√°fico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='product_line', y='gross_income', data=df_supermarket_sales, ci=None)

# A√±adir etiquetas y t√≠tulo
plt.xlabel('L√≠nea de Producto')
plt.ylabel('Ingreso Bruto')
plt.title('Ingreso Bruto por L√≠nea de Producto')

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45, ha='right')

# Mostrar el gr√°fico
plt.show()

# Crear una tabla de resumen
table_summary = pd.pivot_table(df_supermarket_sales, values='gross_income', index='product_line', aggfunc='sum')
table_summary['Percentage'] = (table_summary['gross_income'] / table_summary['gross_income'].sum()) * 100
table_summary = table_summary.round(2)
table_summary
```

```python
# Agrupar por l√≠nea de producto y calcular la suma de los ingresos brutos
grouped_data = df_supermarket_sales.groupby('product_line')['gross_income'].sum()

# Crear un gr√°fico de torta
plt.figure(figsize=(8, 8))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%', startangle=140)

# A√±adir t√≠tulo
plt.title('Distribuci√≥n del Ingreso Bruto por L√≠nea de Producto')

# Mostrar el gr√°fico
plt.show()
```
